```python
def brighten(AR,BR):
    AH = RGBtoHSV(AR)
    BH = RGBtoHSV(BR)
    l = AH.shape[0]
    b = AH.shape[1]
    
    for i in range(l):
        for j in range(b):
            BH[i][j][2] = AH[i][j][2]
    Final = HSVtoRGB(BH)
    return Final
```

```python
def main(path,alpha,beta,gama):
    Img1 = cv2.imread("drive/My Drive/Images_for_Adarsh/DSC_6139.JPG")
    #cv2_imshow(Img1)
    Img2 = cv2.imread(path)
    hsv1 = cv2.cvtColor(Img1, cv2.COLOR_BGR2HSV)
    hsv2 = cv2.cvtColor(Img2, cv2.COLOR_BGR2HSV)
    l = hsv1.shape[0]
    b = hsv1.shape[1]
    diffv = hsv1 - hsv2
    sumv = 0
    sumb = 0
    sums = 0
    count = 0
    for i in range(l):
        for j in range(b):
            sumb = sumb + Img2[i][j][0]
            sumv = sumv + diffv[i][j][2]
            sums = sums + diffv[i][j][1]
            count = count+1
    a = sumv/count
    x = sumb/count
    s = sums/count
    for i in range(l):
        print(i)
        for j in range(b):
            hsv2[i][j][2] = np.minimum(255, np.maximum((alpha*hsv2[i][j][2]) + (beta*a), 0))
            hsv2[i][j][1] = hsv2[i][j][1] - (gama*s)
    final = cv2.cvtColor(hsv2, cv2.COLOR_HSV2BGR)
    for i in range(l):
        print(i)
        for j in range(b):
            if Img2[i][j][0] > (0.75*x):  
                  final[i][j] = Img2[i][j]
       
    cv2.imwrite("drive/My Drive/Results final1/result-a=1,b=0.5.JPG", final)
```
