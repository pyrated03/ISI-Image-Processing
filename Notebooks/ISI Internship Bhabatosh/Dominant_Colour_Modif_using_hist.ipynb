```
import numpy as np
from PIL import Image
import cv2
from google.colab.patches import cv2_imshow
```

```
def histequal(path):
    Img1 = cv2.imread(path)
    temp = cv2.cvtColor(Img1, cv2.COLOR_BGR2YUV)
    temp[:,:,0] = cv2.equalizeHist(temp[:,:,0])
    histeq = cv2.cvtColor(temp, cv2.COLOR_YUV2BGR)
    cv2.imwrite("drive/My Drive/Results final2/original.JPG", Img1)
    cv2.imwrite("drive/My Drive/Results final2/equalized.JPG", histeq)
```

```
histequal("drive/My Drive/Images_for_Adarsh/DSC_0244.JPG")
```
